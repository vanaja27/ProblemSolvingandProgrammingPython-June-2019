```python
# Comprehensions

keyfile='DataFiles/key.txt'

def dictionarykeyfile(keyfile):
    key={}
    with open(keyfile,'r')as f:
        for line in f:
            line=line.split()
            key[line[0]]=line[1]
    return key

keyfile='DataFiles/key.txt'
key=dictionarykeyfile(keyfile)

evenkeys={item for item in key.items() if int(item[0])%2==0}
evenkeys
```
